```python
import sqlite3
import os
import shutil

ruta_base = "../../01_data/BBDD_Normativa_UPV"

# Conectar a la base de datos
conn = sqlite3.connect('../../01_data/Normativa_SQL.db')
cursor = conn.cursor()

# Obtener las rutas de los PDFs
cursor.execute("SELECT Ruta_Doc FROM tblRutaActuales")
rutas_pdfs = [os.path.join(ruta_base, fila[0].lstrip("/")).replace("\\", "/") for fila in cursor.fetchall()]
```

```python
ruta_a_cambiar = "../../01_data/BBDD_Normativa_UPV/2022/2023202300143600001757.pdf"

# Si la ruta está en la lista, modificarla
if ruta_a_cambiar in rutas_pdfs:
    index = rutas_pdfs.index(ruta_a_cambiar)  # Obtener su índice
    carpeta, archivo = os.path.split(ruta_a_cambiar)  # Separar carpeta y archivo
    archivo_modificado = archivo.replace("2023", "2022", 1)  # Cambiar solo la primera aparición de "2023"
    rutas_pdfs[index] = os.path.join(carpeta, archivo_modificado)  # Reemplazar en la lista
```

```python
# Carpeta destino
carpeta_destino = "../../01_data/pdf_actuales/"

# Crear la carpeta si no existe
os.makedirs(carpeta_destino, exist_ok=True)

# Copiar los archivos
for ruta in rutas_pdfs:
    if os.path.isfile(ruta):
        shutil.copy(ruta, carpeta_destino)
        print(f"Copiado: {ruta}")
    else:
        print(f"No encontrado: {ruta}")
```
